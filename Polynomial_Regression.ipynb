```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("C:/Users/naren/Jupyter Notebook/Position_Salaries.csv")
```

```python
dataset
```

```python
x=dataset.iloc[:,1:-1].values
y=dataset.iloc[:,-1].values
```

```python
print(x)
```

```python
print(y)
```

```python
#As there is only few data within the dataset so it is not necessary to divide the given data set into traning and test sets
from sklearn.linear_model import LinearRegression
Lin_reg=LinearRegression()
Lin_reg.fit(x,y) #Training a linear regression model
```

```python
#Now in order to train a polynomial linear regression
from sklearn.preprocessing import PolynomialFeatures
pol=PolynomialFeatures(degree=6)
x_pol=pol.fit_transform(x)
```

```python
print(x_pol)
```

```python
#training a polynomial linear regressor
from sklearn.linear_model import LinearRegression
Pol_reg=LinearRegression()
Pol_reg.fit(x_pol,y)
```

```python
#Visualizing both linear and Polynomial regression
plt.scatter(x,y,color='red')
plt.plot(x,Lin_reg.predict(x),color='blue')
plt.title('Truth vs Bluff')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```

```python
plt.scatter(x,y,color='red')
plt.plot(x,Pol_reg.predict(x_pol),color='blue')
plt.title('Truth vs Bluff')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```
